```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
alpha = 1
beta = 0.5

# Function definitions
def cost_unit(C_t, alpha, beta):
    return alpha * C_t**(-beta)

def cost_total(C_t, alpha, beta):
    return (alpha / (1 - beta)) * C_t**(-beta + 1)

# Values for C_t
C_t_values = np.linspace(0.1, 10, 400)

# Compute the functions
cost_unit_values = cost_unit(C_t_values, alpha, beta)
cost_total_values = cost_total(C_t_values, alpha, beta)

# Plotting the functions
plt.figure(figsize=(9.5, 4))

plt.subplot(1, 2, 1)
plt.plot(C_t_values, cost_unit_values, label=r'$CostUnit = \alpha \cdot C_t^{-\beta}$')
plt.xlabel(r'$Built Capacity(C_t:TW)$')
plt.ylabel(r'$Capital Cost per Unit\ (B\$)$')
plt.title(r'$CostUnit$ vs $Built Capacity(C_t)$')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(C_t_values, cost_total_values, label=r'$CostTotal = \frac{\alpha}{1 - \beta} \cdot C_t^{-\beta + 1}$')
plt.xlabel(r'$Built Capacity(C_t:TW)$')
plt.ylabel(r'$Total Capital Cost\ (B\$)$')
plt.title(r'$CostTotal$ vs $Built Capacity(C_t)$')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
alpha = 1.0
beta = 0.5

def cost_total(C_t, alpha, beta):
    return (alpha / (1 - beta)) * C_t**(-beta + 1)

# Values for C_t
C_t_values_linear = np.linspace(0.1, 10, 400)
C_t_values_non_linear = np.linspace(0.1, 10, 5)

# Compute the functions
cost_total_values_linear = cost_total(C_t_values_linear, alpha, beta)
cost_total_values_non_linear = cost_total(C_t_values_non_linear, alpha, beta)
non_linear_points_on_linear = cost_total(C_t_values_non_linear, alpha, beta)

# Plotting the functions
plt.figure(figsize=(9.5, 4))

plt.subplot(1, 2, 1)
line1, = plt.plot(C_t_values_linear, cost_total_values_linear, label=r'$CostTotal = \frac{\alpha}{1 - \beta} \cdot C_t^{-\beta + 1}$ (Non-linear)')
plt.scatter(C_t_values_non_linear, non_linear_points_on_linear, color=line1.get_color(), zorder=5)
plt.xlabel(r'$BuiltCapacity(C_t:TW)$')
plt.ylabel(r'$CostTotal\ (B\$)$')
plt.title(r'$CostTotal$ vs $C_t$ (Non-linear)')
plt.legend(loc='lower right')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(C_t_values_non_linear, cost_total_values_non_linear, 'o-', label=r'$CostTotal = \frac{\alpha}{1 - \beta} \cdot C_t^{-\beta + 1}$ (Linear)')
plt.xlabel(r'$BuiltCapacity(C_t:TW)$')
plt.ylabel(r'$CostTotal\ (B\$)$')
plt.title(r'$CostTotal$ vs $C_t$ (Linear)')
plt.legend(loc='lower right')
plt.grid(True)

plt.tight_layout()
plt.show()
```
