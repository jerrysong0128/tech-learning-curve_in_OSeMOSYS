```python
import csv

def save_result(output_file_name):
    file_path           = '../export/BTSelResults/'
    output_file_path    = file_path + output_file_name
    # Open the input file in read mode and output file in write mode
    with open('./BTSelResults.csv', 'r') as infile, open(output_file_path, 'w', newline='') as outfile:
        # CSV reader and writer
        reader = csv.reader(infile)
        writer = csv.writer(outfile)

        for row in reader:
            # Skip empty rows
            if not row:
                continue

            # Add empty strings to the row until it has 35 columns
            while len(row) < 36:
                row.append('')
            writer.writerow(row)
```

```python
save_result("Scenario_A_BAU.csv")
```
